La solución de la ecuación:

$$\dfrac{\partial^{2} u}{\partial x^{2}} = \dfrac{\partial^{2} u}{\partial y^{2}} = 0$$ 

$$0<x<1,$$ $$0<y<1$$

sujeta a las condiciones de frontera:

$$a) u(0,y) = u(1,y) = 0;u(x,0) = 100;u(x,1) = 200$$
$$b) u(x,y) = 10y; \dfrac{\partial u}{\partial x} \mid_{x=1} = -1;u(x,0) = u(x,1)=0$$

Esta dada en términos de la serie de Fourier por la expresión:

$$\Sigma_{n=1}^{\infty} = [\dfrac{-20(-1)^{n}}{n\pi} cosh(n\pi x)+[\dfrac{-2(1-(-1)^{n})+20(-1)^{n}coh(n\pi)}{n\pi}]senh(n\pi x)]sen(n\pi y)$$




```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

Usamos una serie de Fourier para aproximar la condición inicial:

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
def u_x_y(x,y):
    N = 8
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = -20*(-1)**n/(n*pi)
          b = (20*(-1)**n*n*pi*sinh(n*pi)-2*(1-(-1)**n))/(n**2*pi**2*cosh(n*pi))
          f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('x'); plt.ylabel('y'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="plasma");
```
