la solucion de la ecuacion:

$$\dfrac{\partial^{2} u}{\partial x^{2}} = \dfrac{\partial^{2} u}{\partial y^{2}} = 0$$ 

$$0<x<1,$$ $$0<y<1$$

sujeta a las condiciones de frontera:

$$a) u(0,y) = u(1,y) = 0;u(x,0) = 100;u(x,1) = 200$$
$$b) u(x,y) = 10y; \dfrac{\partial u}{\partial x} \mid_{x=1} = -1;u(x,0) = u(x,1)=0$$

Esta dada en términos de la serie de Fourier por la expresión:

$u(x,y) = [\dfrac{200[1-(-1)^{n}]}{n\pi } cosh(n\pi y) + \dfrac{200[1-(-1)^{n}][2-cosh(n\pi)]}{senh(n\pi)(n\pi)^{2}} senh(n\pi y)] sen(n\pi x)$

```python
from mpl_toolkits.mplot3d import Axes3D
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
def u_x_y(x,y):
    N = 5
    f = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          a = (200*(1+(-1)^(n+1)))/(n*pi)
          b = (200*(1+(-1)^(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += sin(n*pi*x)*(a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))
    return f


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="viridis");
 
```
